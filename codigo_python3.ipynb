# Solución numérica de la Ecuación de Lane-Emden

```python
from sympy import *
import math
import numpy as np
import matplotlib.pyplot as plt
```

### La solución en serie a la ecuación de Lane-Emden asume la forma

$y = 1+\sum_{k=0}^{\infty}a_{k}x^{2k}$ 

donde los coeficientes $a_k$ estan dados por

$a_{k+1}=\frac{1}{k(k+1)(2k+3)}\sum_{i=1}^{k}(in+i-k)(k-i+1)(3+2k-2i)a_{i}a_{k-i+1}, \qquad k\geq 1$

donde $a_1=\frac{1}{6}x^2$

### Calculo de los coeficientes $a_k$

```python
def coef(a,k,n):
    aux = 1/(k*(k+1)*(2*k+3))
    coeficientes = 0
    for i in np.arange(1,k+1):
        coeficientes = coeficientes + (i*n+i-k)*(k-i+1)*(3+2*k-2*i)*a[i]*a[k-i+1]
    return np.append(a,aux*coeficientes)    
```

# Solucion numerica: iteraciones 

```python
def sol(n,p,x):
    sol1 = 0
    a = np.array([0, -1/6])
    for k in np.arange(0,p):
        sol1 = sol1 + a[k]*(x**(2*k))
        if (k>=1):
            a = coef(a,k,n)  
    return 1 + sol1
```
